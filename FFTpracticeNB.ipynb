```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.fft import fft, ifft
%matplotlib inline
```

```python
plt.style.use('seaborn-poster')
```

```python
sample_rate = 137 # Hz
sample_interval = 1.0/sample_rate # seconds
t = np.arange(0, 1, sample_interval)
```

```python
x = t*0
for frequency, amplitude in ((1, 3), (4, 1), (7, 0.5)):
    x += amplitude * np.sin(2*np.pi*frequency*t)

for frequency, amplitude in ((3, 3), (5, 1), (9, 0.5)):
    x += amplitude * np.cos(2*np.pi*frequency*t)

# x = 3sin(2pi * t) + sin(2pi * 4t) + 0.5sin(2pi * 7t)
```

```python
plt.figure(figsize = (5,5))
plt.plot(t, x, 'r')
plt.ylabel('Amplitude')
plt.show()
```

```python
fft_result = fft(x)
fft_length = len(fft_result)
magnitudes = np.abs(fft_result) 
n = np.arange(fft_length)
frequencies = n/(fft_length/sample_rate)

plt.figure(figsize=(5,5))
plt.subplot(121)
plt.stem(frequencies, magnitudes, 'b', \
         markerfmt=" ", basefmt="-b")
plt.xlabel("Freq (Hz)")
plt.ylabel("Amplitude |X(freq)|")
plt.xlim(-1, 15)
plt.show()
```
